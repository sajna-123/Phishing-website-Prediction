```python
def continuos_univariate_analysis(data1, feature, figsize=(12, 7), kde = False):
    f1, (ax_box1, ax_hist1) = plt.subplots(nrows=2, sharex=True, gridspec_kw={'height_ratios':(0.25, 0.75)}, figsize=figsize)
    sns.color_palette('Pastel1')
    sns.boxplot(data1=data1, x=feature, ax = ax_box1, showmeans=True, color='Violet')
    sns.histplot(data1=data1, x=feature, ax=ax_hist1, kde=kde)
    ax_hist1.axvline(data1[feature].mean(), color='green', linestyle='--')
    ax_hist1.axvline(data1[feature].median(), color='orange', linestyle='-')
    
```
